Loading the dataset using url provided and creating a dataframe

```python
#to load the dataset provided 
#i will use read_csv to read the URl provided
df=pd.read_csv("http://bit.ly/autolib_dataset")
#accesing the dataframe provided to check for the shape
df.shape

```

cleaning the dataset

checking null values and droping them using dropna()

```python
#Access the datset to check jfor leading spaces according to observation there are no leading spaces in this datframe
#check for typing errors and fix them
#checking for missing values in the datframe as a whole
df.isnull()
#checking missig values across each column
df.isnull().any()
#checking how many missing cvalues are there hacross each column
df.isnull().sum()
#dropping the nan values
df1=df.dropna(axis=1)
df1.head(10)

```

checking for duplicates and droping them

```python
#identifying which observations are non duplicates

df_non_duplicates=df1.drop_duplicates()
df_non_duplicates.head(10)
```

dealing with ouliers


using box plot to check for outliers

```python
# Visualising outliers using box plot.
# ---
# A boxplot is a method for graphically depicting groups of numerical data 
# through their quantiles or the five-number summary 
# (minimum, first quartile (Q1), median, third quartile (Q3), and maximum). 
# This method is normally used to detect outliers. 
#importing the liblaries 
import matplotlib.pyplot as plt 
%matplotlib inline 
df_non_duplicates.boxplot(column =['Utilib counter'],grid = False) 
```

```python
#use histogram to visualize the diustribution of the numerical value
df_non_duplicates.Cars.hist()
```

```python
# The scatterplot will visualize the relationship between two quantitative variables. 
# The data points will be displayed as a collection of points. 
# Data points outside the general clustering of the two variables can be indicated as outliers.
fig, ax = plt.subplots(figsize=(12,6))
ax.scatter(df_non_duplicates['Cars'], df['Utilib counter'])
ax.set_xlabel('Cars')
ax.set_ylabel('Utilib countr')
plt.show()
```

Dropping outliers

```python
#droping outliers
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
# Removing outliers based on the IQR range and stores the result in the data frame 'df_out'
dfs = df_non_duplicates[~((df_non_duplicates < (Q1 - 1.5 * IQR)) | (df_non_duplicates > (Q3 + 1.5 * IQR))).any(axis=1)]
#printing the shape of the new datframe

print(dfs.shape)
#printing shape of the old datframe
print(df_non_duplicates.shape)


```
